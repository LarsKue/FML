```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weights = np.load('weights_updates.npy')
print(weights.shape)
print(np.round(weights[-1], 3))
```

```python
feature_names = [
    'move towards coin',
    'collect coin',
    'invalid move',
    'move out of explosion',
    'stay/move in explosion',
    'move towards crate',
    'bomb is suicidal',
    'bomb will destroy crates, if not suicidal',
    'go into dead end if prev move was BOMB',
    'move towards enemy',
    'place bomb if standing next to enemy',
    'place bomb if enemy in explosion area'
]
plt.figure(figsize=(13,13))
for i in range(weights.shape[1]):
    plt.plot(np.arange(200), weights[:200,i], label=feature_names[i])
    
plt.legend()
```
